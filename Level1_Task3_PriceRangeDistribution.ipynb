```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

os.chdir("C:\\Users\\Sarthak\\.jupyter")
restaurants_df = pd.read_csv("File.csv")

# Unique values and their counts in the 'Price Range' column
price_range_counts = restaurants_df['Price range'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
price_range_counts.plot(kind='bar', color='brown')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price range')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Percentage of restaurants in each price range category
total_restaurants = len(restaurants_df)
percentage_in_each_category = (price_range_counts / total_restaurants) * 100
print("Percentage of restaurants in each price range category:")
print(percentage_in_each_category)
```
